# ML models

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.metrics import log_loss, confusion_matrix
from sklearn.linear_model import SGDClassifier
from sklearn.calibration import CalibratedClassifierCV
import xgboost as xgb
import warnings
warnings.filterwarnings("ignore")
```

## Checking data

```python
data = pd.read_csv('final_features_idf_w2v.csv')
```

```python
data.head()
```

```python
data[['1_x','95_x','1_y','95_y']].info()
```

```python
data[data.columns[:29]].info()
```

```python
data = data[:100000]
y_true = data['is_duplicate']
data.drop(['id','is_duplicate'],axis=1,inplace=True)
```

## Random train test split (70:30)

```python
x_train, x_test, y_train, y_test = train_test_split(data,y_true,stratify=y_true,test_size=0.3)
```

```python
print("Number of data points in train data :",x_train.shape)
print("Number of data points in test data :",x_test.shape)
```

```python
print('-'*10,'Distribution of output variable in train data','-'*10)
train_distr = Counter(y_train)
print('Class 0: ',train_distr[0]/len(y_train),' Class 1: ',train_distr[1]/len(y_train))
print('-'*10,'Distribution of output variable in test data','-'*10)
test_distr = Counter(y_test)
print('Class 0: ',test_distr[0]/len(y_test),' Class 1: ',test_distr[1]/len(y_test))
```

## Function to plot confusion matrices

```python
def plot_confusion_matrix(test_y,pred_y):
    C = confusion_matrix(test_y,pred_y)
    A =(((C.T)/(C.sum(axis=1))).T)
    B =(C/C.sum(axis=0))
    
    plt.figure(figsize=(20,4))
    labels=[0,1]
    
    cmap = sns.color_palette("Blues")
    plt.subplot(1,3,1)
    sns.heatmap(C,annot=True,cmap=cmap,fmt=".3f",xticklabels=labels,yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Confusion matrix")
    
    plt.subplot(1,3,2)
    sns.heatmap(B,annot=True,cmap=cmap,fmt=".3f",xticklabels=labels,yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Precision matrix")
    
    plt.subplot(1,3,3)
    sns.heatmap(A,annot=True,cmap=cmap,fmt=".3f",xticklabels=labels,yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Recall matrix")
    
    plt.show()
```

## Random model (For worst case log loss)

```python
predicted_y = np.zeros((len(y_test),2))

for i in range(len(y_test)):
    prob = np.random.rand(1, 2)
    predicted_y[i] = (prob/sum(sum(prob)))[0]
    
print("Log loss on Test Data using Random Model",log_loss(y_test, predicted_y, eps=1e-15))

predicted_y = np.argmax(predicted_y, axis=1)
plot_confusion_matrix(y_test, predicted_y)
```

## Logistic regression
### Hyperparameter tuning

```python
# hyperparameter for SGD Classifier
# alpha ranges from 0.00005 to 10 (7 values)
alpha = [10**x for x in range(-5, 2)]
log_error_array = []

for i in alpha:
    clf = SGDClassifier(alpha=i,penalty='l2',loss='log',random_state=42)
    clf.fit(x_train,y_train)
    sig_clf = CalibratedClassifierCV(clf,method='sigmoid')
    sig_clf.fit(x_train,y_train)
    predicted_y = sig_clf.predict_proba(x_test)
    log_error_array.append(log_loss(y_test,predicted_y,labels=clf.classes_,eps=1e-15))
    print("Alpha: ",i," Log loss: ",log_loss(y_test,predicted_y,labels=clf.classes_,eps=1e-15))

fig, ax = plt.subplots()
ax.plot(alpha,log_error_array)
for i,txt in enumerate(np.round(log_error_array,3)):
    ax.annotate((alpha[i],np.round(txt,3)),(alpha[i],log_error_array[i]))
plt.grid()
plt.title("Cross Validation Error for each alpha")
plt.xlabel("Alpha i's")
plt.ylabel("Error measure")
plt.show()
```

### Training and predicting using best parameter

```python
best_alpha = np.argmin(log_error_array)
clf = SGDClassifier(alpha=alpha[best_alpha],penalty='l2',loss='log',random_state=42)
clf.fit(x_train,y_train)
sig_clf = CalibratedClassifierCV(clf,method='sigmoid')
sig_clf.fit(x_train,y_train)

predict_y = sig_clf.predict_proba(x_train)
print('Log loss for best alpha on train data: ',log_loss(y_train,predict_y,labels=clf.classes_, eps=1e-15))
predict_y = sig_clf.predict_proba(x_test)
print('Log loss for best alpha on test data: ',log_loss(y_test,predict_y,labels=clf.classes_, eps=1e-15))
predicted_y = np.argmax(predict_y,axis=1)
print('Total data points in test: ',len(predicted_y))
plot_confusion_matrix(y_test,predicted_y)
```

## Linear SVM
### Hyperparameter tuning

```python
# hyperparameter for SGD Classifier
# alpha ranges from 0.00005 to 10 (7 values)
alpha = [10**x for x in range(-5, 2)]
log_error_array = []

for i in alpha:
    clf = SGDClassifier(alpha=i,penalty='l2',loss='hinge',random_state=42)
    clf.fit(x_train,y_train)
    sig_clf = CalibratedClassifierCV(clf,method='sigmoid')
    sig_clf.fit(x_train,y_train)
    predicted_y = sig_clf.predict_proba(x_test)
    log_error_array.append(log_loss(y_test,predicted_y,labels=clf.classes_,eps=1e-15))
    print("Alpha: ",i," Log loss: ",log_loss(y_test,predicted_y,labels=clf.classes_,eps=1e-15))

fig, ax = plt.subplots()
ax.plot(alpha,log_error_array)
for i,txt in enumerate(np.round(log_error_array,3)):
    ax.annotate((alpha[i],np.round(txt,3)),(alpha[i],log_error_array[i]))
plt.grid()
plt.title("Cross Validation Error for each alpha")
plt.xlabel("Alpha i's")
plt.ylabel("Error measure")
plt.show()
```

### Training and predicting using best parameter

```python
best_alpha = np.argmin(log_error_array)
clf = SGDClassifier(alpha=alpha[best_alpha],penalty='l2',loss='hinge',random_state=42)
clf.fit(x_train,y_train)
sig_clf = CalibratedClassifierCV(clf,method='sigmoid')
sig_clf.fit(x_train,y_train)

predict_y = sig_clf.predict_proba(x_train)
print('Log loss for best alpha on train data: ',log_loss(y_train,predict_y,labels=clf.classes_, eps=1e-15))
predict_y = sig_clf.predict_proba(x_test)
print('Log loss for best alpha on test data: ',log_loss(y_test,predict_y,labels=clf.classes_, eps=1e-15))
predicted_y = np.argmax(predict_y,axis=1)
print('Total data points in test: ',len(predicted_y))
plot_confusion_matrix(y_test,predicted_y)
```
